```python
import torch
from transformers import MarianMTModel, MarianTokenizer

# Load pre-trained English-to-Swahili model
model_name = "Helsinki-NLP/opus-mt-en-sw"
tokenizer = MarianTokenizer.from_pretrained(model_name)
model = MarianMTModel.from_pretrained(model_name)

def translate(text, tokenizer, model):
    """ Translates English text into Swahili using MarianMT. """
    inputs = tokenizer(text, return_tensors="pt", padding=True, truncation=True)
    with torch.no_grad():
        output = model.generate(**inputs)
    return tokenizer.decode(output[0], skip_special_tokens=True)

# Interactive user input
print("🔠 English-to-Swahili Translator (Type 'exit' to quit) 🔠")
while True:
    text = input("\nEnter English text: ")
    if text.lower() == "exit":
        print("Goodbye! 👋")
        break
    translated_text = translate(text, tokenizer, model)
    print(f"Swahili Translation: {translated_text}")
```
